```python
from TTS.api import TTS
from diffusers import StableDiffusionPipeline
from moviepy.editor import *
from IPython.display import Audio, Image, Video, display
import torch

# ----- INPUT PARAGRAPH -----
paragraph = "The sun set over the quiet hills, painting the sky with hues of orange and pink. A gentle breeze whispered through the trees."

# ----- STEP 1: TEXT TO SPEECH -----
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False)
tts.tts_to_file(text=paragraph, file_path="output_audio.wav")
print("✅ Audio created")

# --- DISPLAY AUDIO ---
display(Audio("output_audio.wav"))

# ----- STEP 2: TEXT TO IMAGE (STABLE DIFFUSION) -----
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

image_prompt = "A scenic sunset over hills, orange and pink sky, trees gently swaying in the wind"
image = pipe(image_prompt).images[0]
image_path = "generated_image.png"
image.save(image_path)
print("✅ Image created")

# --- DISPLAY IMAGE ---
display(Image(filename=image_path))

# ----- STEP 3: MERGE AUDIO + IMAGE INTO VIDEO -----
audio_clip = AudioFileClip("output_audio.wav")
image_clip = ImageClip(image_path).set_duration(audio_clip.duration)
image_clip = image_clip.set_audio(audio_clip)
video_clip = image_clip.resize(height=720).set_fps(24)

video_clip.write_videofile("final_video.mp4", codec="libx264", audio_codec="aac")
print("✅ Video saved as final_video.mp4")

# --- DISPLAY VIDEO ---
display(Video("final_video.mp4", embed=True))
```

```python
# 📥 Step 2: Import libraries
from modelscope.pipelines import pipeline
from modelscope.outputs import OutputKeys
from moviepy.editor import VideoFileClip, AudioFileClip
from TTS.api import TTS
import IPython.display as display

# 📝 Step 3: Input paragraph
text = "The sun sets over the hills as the sky turns orange. A gentle breeze rustles the trees."

# 🎥 Step 4: Generate video using the most used model
video_pipe = pipeline('text-to-video-synthesis', model='Wan-AI/Wan2.1-T2V-14B', device='cuda')
video_output = video_pipe({'text': text})
video_path = video_output[OutputKeys.OUTPUT_VIDEO]
print("✅ Video generated:", video_path)

# 🔊 Step 5: Generate audio using Text-to-Speech
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DD C", progress_bar=False)
tts.tts_to_file(text=text, file_path="tts_audio.wav")
print("✅ Audio generated")

# ▶️ Play generated audio in notebook
display.display(display.Audio("tts_audio.wav"))

# 🎞️ Step 6: Merge audio and video
video = VideoFileClip(video_path)
audio = AudioFileClip("tts_audio.wav").subclip(0, video.duration)
final_video = video.set_audio(audio)
final_video.write_videofile("final_result.mp4", codec="libx264", audio_codec="aac")
print("✅ Final video created")

# ▶️ Step 7: Display final video in Colab
display.display(display.Video("final_result.mp4", embed=True))
```
