```python


# STEP 2: Main App
import gradio as gr
from googlesearch import search
import requests
from bs4 import BeautifulSoup
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from TTS.api import TTS

# Load summarization pipeline
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# Load NLLB model (non-fast tokenizer)
model_name = "facebook/nllb-200-1.3B"
nllb_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
nllb_tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)

# Language map: display -> (lang_code, tts_model)
language_map = {
    "Telugu": ("tel_Telu", "tts_models/te/ekstep/fastspeech2"),
    "Hindi": ("hin_Deva", "tts_models/hi/cv/fastpitch"),
    "Tamil": ("tam_Taml", "tts_models/ta/ekstep/fastspeech2"),
    "Kannada": ("kan_Knda", "tts_models/kn/ekstep/fastspeech2"),
    "English": ("eng_Latn", "tts_models/en/ljspeech/tacotron2-DDC")
}

def translate_with_nllb(text, target_lang_code):
    tokenizer = nllb_tokenizer
    model = nllb_model
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    # Use manually mapped bos_token_id
    inputs['forced_bos_token_id'] = tokenizer.convert_tokens_to_ids(target_lang_code)
    translated_tokens = model.generate(**inputs, max_length=512)
    return tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

def summarize_translate_speak(topic, language):
    try:
        tgt_lang_code, tts_model_name = language_map[language]

        # Step 1: Scrape content from Google results
        results = list(search(topic, num_results=5))
        content = ""
        for url in results:
            try:
                r = requests.get(url, timeout=5)
                soup = BeautifulSoup(r.content, "html.parser")
                paragraphs = soup.find_all("p")
                text = " ".join(p.get_text() for p in paragraphs)
                if len(text.strip()) > 100:
                    content += text + "\n"
            except:
                continue

        if not content:
            return None, "❌ No valid content found from search results."

        content = content[:3000]  # Clip long content
        summary = summarizer(content, max_length=150, min_length=30, do_sample=False)[0]["summary_text"]

        # Step 2: Translate summary
        translated = translate_with_nllb(summary, tgt_lang_code)

        # Step 3: TTS
        tts = TTS(model_name=tts_model_name, progress_bar=False, gpu=False)
        filename = f"{language.lower().replace(' ', '_')}_summary.wav"
        tts.tts_to_file(text=translated, file_path=filename)

        return filename, f"✅ {language} Translation:\n\n{translated}"

    except Exception as e:
        return None, f"❌ Error occurred: {str(e)}"

# UI
with gr.Blocks() as demo:
    gr.Markdown("## 🌐 Multilingual Topic Summarizer with Audio (NLLB + TTS)")

    topic_input = gr.Textbox(label="🔍 Enter a topic (in English)", placeholder="e.g., Water crisis in India")
    language_dropdown = gr.Dropdown(choices=list(language_map.keys()), value="Telugu", label="🌐 Select Output Language")
    generate_button = gr.Button("🎙️ Generate Audio Summary")

    output_audio = gr.Audio(label="🔊 Play Summary", type="filepath")
    output_text = gr.Textbox(label="📝 Translated Summary Text")

    generate_button.click(fn=summarize_translate_speak, inputs=[topic_input, language_dropdown], outputs=[output_audio, output_text])

demo.launch(debug=True)
```

```python
import gradio as gr
from googlesearch import search
import requests
from bs4 import BeautifulSoup
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer, pipeline
from TTS.api import TTS
import tempfile

# Summarizer
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# Translator
m2m_model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
m2m_tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")

# Language map: (lang_code, TTS model)
language_map = {
    "Telugu": ("te", "tts_models/te/ekstep/fastspeech2"),
    "Hindi": ("hi", "tts_models/hi/ekstep/vakyansh"),
    "Tamil": ("ta", "tts_models/ta/ekstep/fastspeech2"),
    "Kannada": ("kn", "tts_models/kn/ekstep/fastspeech2"),
    "Marathi": ("mr", "tts_models/mr/cv/bert_vits2"),
    "Gujarati": ("gu", "tts_models/gu/cv/bert_vits2"),
    "Bengali": ("bn", "tts_models/bn/cv/bert_vits2"),
    "Malayalam": ("ml", "tts_models/ml/cv/bert_vits2"),
    "Urdu": ("ur", "tts_models/ur/cv/bert_vits2")
}

# Translation function
def translate(text, src_lang, tgt_lang):
    m2m_tokenizer.src_lang = src_lang
    encoded = m2m_tokenizer(text, return_tensors="pt")
    generated = m2m_model.generate(**encoded, forced_bos_token_id=m2m_tokenizer.get_lang_id(tgt_lang))
    return m2m_tokenizer.batch_decode(generated, skip_special_tokens=True)[0]

# Core pipeline
def summarize_translate_speak(topic, language):
    try:
        tgt_lang_code, tts_model_name = language_map[language]

        # Step 1: Web scraping
        results = list(search(topic, num_results=5))
        content = ""
        for url in results:
            try:
                res = requests.get(url, timeout=5)
                soup = BeautifulSoup(res.content, "html.parser")
                paragraphs = soup.find_all("p")
                content += " ".join(p.get_text() for p in paragraphs) + "\n"
            except:
                continue

        # Step 2: Summarize
        content = content[:3000]
        summary_en = summarizer(content, max_length=150, min_length=30, do_sample=False)[0]["summary_text"]

        # Step 3: Translate
        summary_translated = translate(summary_en, "en", tgt_lang_code)

        # Step 4: TTS - English
        tts_en = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)
        with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as f_en:
            tts_en.tts_to_file(text=summary_en, file_path=f_en.name)
            english_audio_path = f_en.name

        # Step 5: TTS - Translated
        tts = TTS(model_name=tts_model_name, progress_bar=False, gpu=False)
        with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as f_trans:
            tts.tts_to_file(text=summary_translated, file_path=f_trans.name)
            translated_audio_path = f_trans.name

        return english_audio_path, summary_en, translated_audio_path, summary_translated

    except Exception as e:
        return None, "❌ Error occurred during English processing.", None, f"❌ Error: {str(e)}"

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🌍 AI Web Summarizer + Dual Audio Translator")

    topic_input = gr.Textbox(label="Enter a Topic (English)", placeholder="e.g., Artificial Intelligence")
    language_dropdown = gr.Dropdown(choices=list(language_map.keys()), value="Hindi", label="Choose Target Language")
    submit_button = gr.Button("🔎 Search, Summarize, Translate & Speak")

    gr.Markdown("### 📢 English Summary")
    output_en_audio = gr.Audio(label="🔊 English Audio", type="filepath")
    output_en_text = gr.Textbox(label="📝 English Summary Text")

    gr.Markdown("### 🌐 Translated Summary")
    output_trans_audio = gr.Audio(label="🔊 Translated Audio", type="filepath")
    output_trans_text = gr.Textbox(label="📝 Translated Summary Text")

    submit_button.click(
        fn=summarize_translate_speak,
        inputs=[topic_input, language_dropdown],
        outputs=[output_en_audio, output_en_text, output_trans_audio, output_trans_text]
    )

demo.launch(debug=True)
```
