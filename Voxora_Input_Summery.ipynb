```python
# STEP 1: INSTALL REQUIRED LIBRARIES
# !pip install -q gradio googlesearch-python transformers TTS torch requests beautifulsoup4

# STEP 2: IMPORTS
import gradio as gr
from googlesearch import search
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
from TTS.api import TTS
import os

# STEP 3: LOAD MODELS
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=False, gpu=False)
default_speaker = tts.speakers[0]

# STEP 4: MAIN LOGIC (ENGLISH SUMMARY AND AUDIO ONLY)
def process_pipeline(topic):
    try:
        # STEP 1: SCRAPE WEB CONTENT
        content = ""
        for url in search(topic, num_results=10):
            try:
                res = requests.get(url, timeout=5)
                soup = BeautifulSoup(res.text, "html.parser")
                content += " ".join([p.text for p in soup.find_all("p")]) + "\n"
            except:
                continue

        content = content[:3000]  # Limit content to avoid overload

        # STEP 2: SUMMARIZE IN ENGLISH
        summary_en = summarizer(content, max_length=150, min_length=40, do_sample=False)[0]['summary_text']

        # STEP 3: TEXT-TO-SPEECH (ENGLISH)
        en_audio_path = "english_summary.wav"
        if summary_en.strip():
            tts.tts_to_file(text=summary_en, speaker=default_speaker, language="en", file_path=en_audio_path)

        return en_audio_path, summary_en

    except Exception as e:
        return None, "Error occurred during processing."

# STEP 5: GRADIO UI
with gr.Blocks() as demo:
    gr.Markdown("## 🗣️ Voxora ")

    topic = gr.Textbox(label="Enter your topic")

    with gr.Row():
        audio_en = gr.Audio(label="🔊 English Audio", type="filepath")
        summary_text = gr.Textbox(label="📝 English Summary")

    btn = gr.Button("Generate Audio")

    btn.click(
        fn=process_pipeline,
        inputs=[topic],
        outputs=[audio_en, summary_text]
    )

demo.launch(debug=True)
```
